```python
import imageio
import imgaug as ia
import imgaug.augmenters as iaa
import ipyplot
import matplotlib.pyplot as plt
```

```python
input_img = imageio.imread('experiments/patches/tile_overlap/F005a02/F005a02_0_0.tif')
```

```python
#Horizontal Flip
hflip= iaa.Fliplr(p=1.0)
input_hf= hflip.augment_image(input_img)
```

```python
#Vertical Flip
vflip= iaa.Flipud(p=1.0)
input_vf= vflip.augment_image(input_img)
```

```python
#Rotate
rflip= iaa.Rotate((270, 270)) 
input_rf= rflip.augment_image(input_img)
```

```python
images_list=[input_img, input_hf, input_vf, input_rf]
labels = ['Original', 'Horizontally flipped', 'Vertically flipped', 'Rotated +270']
ipyplot.plot_images(images_list,labels=labels,img_width=180)
```
